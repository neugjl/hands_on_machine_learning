```python
from scipy.io import wavfile
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data_path ="N:/Doopler/matlab_functions/BUG20171220/data/NX606Jautolabel5/"
data_path ="./"
file_name = "96K16bit2.wav"
fs,data = wavfile.read(data_path+file_name,mmap=False)
points=np.arange(data.size)
times = points/fs
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(points,data,'r')
axes[1].plot(times,data,'b')
fig.tight_layout()
```
